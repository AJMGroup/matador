```python
from matador import DBQuery, QueryConvexHull
query = DBQuery(composition=['NaP'], summary=True, subcmd='hull', db='ajm')
```

```python
hull = QueryConvexHull(query, no_plot=False, hull_cutoff=26e-3)
```
