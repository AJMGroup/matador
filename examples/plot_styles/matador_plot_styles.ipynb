# Plotting in matador

This notebook shows how to alter plotting styles in matador. Typically, these results would be achieved by editing the `default_style` field in your `.matadorrc` file.

```python
%matplotlib inline
from matador.query import DBQuery
from matador.hull import QueryConvexHull
from matador.scrapers import res2dict
import glob
import os
from matador.plotting import plot_2d_hull
from matador.plotting import plot_spectral

# change this if you are running from outside the matador/examples folder
MATADOR_SRC_DIR = '../../../'
```

```python
cursor = res2dict(glob.glob(MATADOR_SRC_DIR + 'matador/tests/data/hull-KP-KSnP_pub/*.res'))
query = DBQuery(cursor=cursor, composition='KP', subcmd='hull')
hull = QueryConvexHull(query, no_plot=True)
```

```python
styles = ['matador', 'seaborn', 'seaborn-whitegrid', 'dark_background', 'default', 'classic']
for style in styles:
    plot_2d_hull(hull, style=style)
    plot_spectral('{}/matador/tests/data/dispersion/K3P-OQMD_4786-CollCode25550'.format(MATADOR_SRC_DIR), 
                  style=style)
```
